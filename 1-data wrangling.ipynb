now combine both datasets into a single dataset

set a column "popularity" which is 1 for high popular songs and 0 for low popular songs

```python
data_high_popularity['popularity'] = 1
data_low_popularity['popularity'] = 0
data = pd.concat([data_high_popularity, data_low_popularity], ignore_index=True)
```

also "track_href", "uri", "analysis_url", "track_album_id", "track_id", "playlist_id" attributes are unique to each entry and can be removed from the dataset. [keeping the "id" feature as a unique identifer for each entry]

```python
data.drop(columns=["playlist_subgenre", "playlist_name", "track_href", "track_id", "type", "playlist_id",
                   "uri", "analysis_url", "track_album_id"], inplace=True)
```

```python
data.head()
```

get the year of release of each track

```python
data['release_date'] = pd.to_datetime(data['track_album_release_date'], errors='coerce')
data.head()
```

```python
data['release_date'].isnull().sum()
```

```python
data.loc[data['release_date'].isnull(), 'track_album_release_date'].unique()
```

```python
data['release_year'] = data['release_date'].dt.year
```

```python
data['release_year'].isnull().sum()
```

fill the null values of "release_year" with data from "track_album_release_date"

```python
data.loc[data['release_year'].isnull(), 'release_year'] = data.loc[data['release_year'].isnull(), 'track_album_release_date'].str[:4]
```

```python
data['release_year'].isnull().sum()
```

```python
data.head()
```

```python
data['release_year'] = data['release_year'].astype(int)
```

```python
data.drop(columns=['track_album_release_date', 'release_date'], inplace=True)
```

```python
data.rename(columns={'track_artist':'artist', 'track_album_name':'album',
             'track_name':'name', 'playlist_genre':'genre'}, inplace=True)
data.head()
```

```python
data.columns[data.isnull().any()]
```

the data set has no null values now

```python
data.duplicated().sum()
```

drop the duplicates

```python
data = data.drop_duplicates()
data.duplicated().sum()
```

the dataset is cleaned and missing values are replaced
